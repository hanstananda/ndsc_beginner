# Initial text analysis
### dic will contain pair of word and its category mapped to its occurence in trainData
### dicAll will contain word mapped its occurence in trainData

```python
dic = {}

for index,row in trainData.iterrows():
    arr = re.split('\W+',row["title"])
    for sz in arr:
        if (sz,row["Category"]) not in dic:
            dic[(sz,row["Category"])] = 1
        else:
            dic[(sz,row["Category"])]+=1

# for i in range(trainData.shape[0]):
#     arr = re.split('\W+',trainData["title"][i])
#     for j in arr:
#         if (j,trainData["Category"][i]) not in dic:
#             dic[(j,trainData["Category"][i])]=1
#         else:
#             dic[(j,trainData["Category"][i])]+=1
```

```python
dicAll = {}


for index,row in trainData.iterrows():
    arr = re.split('\W+',row["title"])
    for sz in arr:
        if sz not in dicAll:
            dicAll[sz] = 1
        else:
            dicAll[sz]+=1
```

```python
idx = []
res = []
```

```python
for index,row in testData.iterrows():
    if index>1000:
        break
    maxi = 0.0
    imax = 0
    for i in range(58):
        tot = 0.0
        arr = re.split('\W+',row["title"])
        for sz in arr:
            if (sz,i) in dic:
                tot += dic[(sz,i)]/dicAll[sz]
        if tot>maxi:
            maxi=tot
            imax = i
    idx.append(row["itemid"])
    res.append(imax)
```

```python
df = pd.DataFrame({'itemid': idx, 'Category':res})
```
