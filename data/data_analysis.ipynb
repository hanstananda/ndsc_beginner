# Initial text analysis
### dic will contain pair of word and its category mapped to its occurence in trainData
### dicAll will contain word mapped its occurence in trainData

```python
dic = {}

for index,row in trainData.iterrows():
    arr = re.split('\W+',row["title"])
    for sz in arr:
        if (sz,row["Category"]) not in dic:
            dic[(sz,row["Category"])] = 1
        else:
            dic[(sz,row["Category"])]+=1

# for i in range(trainData.shape[0]):
#     arr = re.split('\W+',trainData["title"][i])
#     for j in arr:
#         if (j,trainData["Category"][i]) not in dic:
#             dic[(j,trainData["Category"][i])]=1
#         else:
#             dic[(j,trainData["Category"][i])]+=1
```

```python
dicAll = {}


for index,row in trainData.iterrows():
    arr = re.split('\W+',row["title"])
    for sz in arr:
        if sz not in dicAll:
            dicAll[sz] = 1
        else:
            dicAll[sz]+=1
```

```python
idx = []
res = []
```

```python
for index,row in testData.iterrows():
    if index>1000:
        break
    maxi = 0.0
    imax = 0
    for i in range(58):
        tot = 0.0
        arr = re.split('\W+',row["title"])
        for sz in arr:
            if (sz,i) in dic:
                tot += dic[(sz,i)]/dicAll[sz]
        if tot>maxi:
            maxi=tot
            imax = i
    idx.append(row["itemid"])
    res.append(imax)
```

```python
df = pd.DataFrame({'itemid': idx, 'Category':res})
```

```python
dic_out ={}
dic_out_pre = {}
dic_out_suf = {}
for index,row in trainData.iterrows():
    arr = re.split('\W+',row["title"])
    for i in range(len(arr)):
        if arr[i].isdigit():
            bef = ""
            aft1 = ""
            aft2 = ""
            if i!=0:
                bef = arr[i-1]
            if i<len(arr)-1:
                aft1 = arr[i+1]
            if i<len(arr)-2:
                aft2 = arr[i+2]
            if (bef,arr[i],aft1,aft2) in dic_out:
                dic_out[(bef,arr[i],aft1,aft2)]+=1
            else:
                dic_out[(bef,arr[i],aft1,aft2)]=1
                
            if bef in dic_out_pre:
                dic_out_pre[bef]+=1
            else:
                dic_out_pre[bef]=1
                
            if (aft1,aft2) in dic_out_suf:
                dic_out_suf[(aft1,aft2)]+=1
            else:
                dic_out_suf[(aft1,aft2)]=1
```

```python
dic_word = {}
for index,row in trainData.iterrows():
    arr = re.split('\W+',row["title"])    
    for sz in arr:
        if re.match('^ter', sz):
            if sz in dic_word:
                dic_word[sz]+=1
            else:
                dic_word[sz]=1
```
